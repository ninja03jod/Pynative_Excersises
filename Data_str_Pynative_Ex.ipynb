```python
# Given two lists, l1 and l2,
# write a program to create a third list l3 by picking an odd-index element from the list l1 and even index elements from the list l2.
l1 = [3, 6, 9, 12, 15, 18, 21]
l2 = [4, 8, 12, 16, 20, 24, 28]
l3 = []
for i in range(1,len(l1),2):
  l3.append(l1[i])

for j in range(0,len(l2),2):
  l3.append(l2[i])
print(l3)
```

```python
l3 = l1[1::2]
# odd index
l3 += l2[0::2]
# even index
print(l3)
```

```python
# Write a program to remove the item present at index 4 and add it to the 2nd position and at the end of the list.
list1 = [54, 44, 27, 79, 91, 41]
# remove item index at 4
list1.remove(91)
print(list1)
# adding to 2nd position
list1.insert(2,91)
print(list1)
# adding at end
list1.append(91)
print(list1)
```

```python
# Slice list into 3 equal chunks and reverse each chunk
sample_list = [11, 45, 8, 23, 14, 12, 78, 45, 89]

ch1 = sample_list[:3]
print(ch1)
print(list(reversed(ch1)))

ch2 = sample_list[3:6]
print(ch2)
print(ch2[::-1])

ch3 = sample_list[6:]
print(ch3)
print(ch3[::-1])
```

```python
# Write a program to iterate a given list and count the occurrence of each element and create a dictionary to show the count of each element.
sample_list = [11, 45, 8, 11, 23, 45, 23, 45, 89]

cnt_dict = dict()
for i in sample_list:
  if i in cnt_dict:
    cnt_dict[i] += 1
  else:
    cnt_dict[i] = 1
print('printing count of each item:',cnt_dict)
```

```python
# Find the intersection (common) of two sets and remove those elements from the first set
first_set = {23, 42, 65, 57, 78, 83, 29}
second_set = {57, 83, 29, 67, 73, 43, 48}

print(first_set.intersection(second_set))

print('First set after removing common element',first_set.difference(second_set))
```

```python
#  Checks if one set is a subset or superset of another set. If found, delete all elements from that set
first_set = {27, 43, 34}
second_set = {34, 93, 22, 27, 43, 53, 48}

print('first_set is subset of second_set',first_set.issubset(second_set))

print('second_set is subset of first_set',second_set.issubset(first_set))

print('first_set is superset of second_set',first_set.issuperset(second_set))

print('second_set is superset of first_set',second_set.issuperset(first_set))

first_set.clear()
print('first set:',first_set)

print('second set:',second_set)
```

```python
# Iterate a given list and check if a given element exists as a keyâ€™s value in a dictionary. If not, delete it from the list
roll_number = [47, 64, 69, 37, 76, 83, 95, 97]
sample_dict = {'Jhon':47, 'Emma':69, 'Kelly':76, 'Jason':97}

for i in roll_number:
  if i in sample_dict.values():
    print(i,end=' ')

print([i for i in roll_number if i in sample_dict.values()])
```
