```python
# Merge two Python dictionaries into one
dict1 = {'Ten': 10, 'Twenty': 20, 'Thirty': 30}
dict2 = {'Thirty': 30, 'Fourty': 40, 'Fifty': 50}

dict3 = dict1
dict3.update(dict2)
print(dict3)

dict3 = {**dict1,**dict2}
print(dict3)
```

```python
# Create a dictionary by extracting the keys from a given dictionary
sample_dict = {
    "name": "Kelly",
    "age": 25,
    "salary": 8000,
    "city": "New york"}

# Keys to extract
keys = ["name", "salary"]
new_dict = {}
new_dict.update({'name':'kelly','salary':8000})
print(new_dict)

ndict = {k:sample_dict[k] for k in keys}
print(ndict)
```

```python
# Delete a list of keys from a dictionary
sample_dict = {
    "name": "Kelly",
    "age": 25,
    "salary": 8000,
    "city": "New york"}

# Keys to extract
keys = ["name", "salary"]

nw_dict = {sample_dict.pop(k) for k in keys}
print(sample_dict)

nedict = {}
nedict.update({'city':'New york','age':25})
print(nedict)
```
