```python
# Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list,
# then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.
list1 = ["M", "na", "i", "Ke"]
list2 = ["y", "me", "s", "lly"]
list3 = []
res1 = list1[0]+list2[0]
list3.append(res1)
res2 = list1[1]+list2[1]
list3.append(res2)
res3 = list1[2]+list2[2]
list3.append(res3)
res4 = list1[3]+list2[3]
list3.append(res4)
print(list3)

# -----------------------------

list3 = [i+j for i, j in zip(list1,list2)]
print(list3)

print([i+j for i,j in zip(list1,list2)])
```
